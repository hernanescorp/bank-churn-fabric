Set ctes 

```python
IS_CUSTOM_DATA = False 

DATA_ROOT = "/lakehouse/default"
DATA_FOLDER = "Files/churn"
DATA_FILE = "churn.csv"
```

Downloading dataset from azure

```python
import os, requests
if not IS_CUSTOM_DATA:
    remote_url = "https://synapseaisolutionsa.z13.web.core.windows.net/data/bankcustomerchurn"
    file_list = [DATA_FILE]
    dowload_path = f"{DATA_ROOT}/{DATA_FOLDER}/raw"

    if not os.path.exists("/lakehouse/default"):
        raise FileNotFoundError(
            "Lakehouse does not exists"
        )
    os.makedirs(dowload_path, exist_ok = True)
    for fname in file_list:
        if not os.path.exists(f"{dowload_path}/{fname}"):
            r = requests.get(f"{remote_url}/{fname}",timeout=30)
            with open(f"{dowload_path}/{fname}", "wb") as f:
                f.write(r.content)
    print("Downloaded demo data")

```
